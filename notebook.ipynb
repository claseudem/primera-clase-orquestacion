```python
from prefect import task, flow
```

```python
@task
def formato_tres_cinco_uno():
    print("Tarea 1")

@task
def filtrar_solo_us(formato):
    print("Tarea 2")

@task
def solo_usd_rv(filtrar):
    print("Tarea 3")

@task
def armar_portafolio(rv_usd):
    print("Tarea 4")

@task
def compras_y_ventas(armar):
    print("Tarea 5")
```
